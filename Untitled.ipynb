```python
import pandas as pd
import matplotlib.pyplot as plt

log_files = {
    "Energy-aware": "Energy-aware matching_response_logs.csv",
    "Deadline-energy-aware": "Deadline-energy-aware matching_response_logs.csv",
    "Multi-criteria": "Multi-criteria matching_response_logs.csv",
    "Hungarian": "Hungarian_response_logs.csv",
    "Random": "Random_response_logs.csv"
}

ax_text = ['a', 'b']

model_counts = {}
node_counts= {}

for approach, file in log_files.items():
    df = pd.read_csv(file)
    counts_model = df["model"].value_counts()
    model_counts[approach] = counts_model

    counts_node = df["edge_node_id"].value_counts()
    node_counts[approach] = counts_node

assigned_model = pd.DataFrame(model_counts).fillna(0).astype(int).sort_index()
assigned_node = pd.DataFrame(node_counts).fillna(0).astype(int).sort_index()

fig, axs = plt.subplots(1, 2, figsize=(20, 5), sharey=False)
assigned_model.plot(kind='bar', ax=axs[0])

axs[0].set_title("Requests per Model by Approach")
axs[0].set_xlabel("(a) Model")
axs[0].set_ylabel("Number of Assigned Requests")
axs[0].legend(title="Approach")
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

assigned_node.plot(kind='bar', ax=axs[1])
axs[1].set_title("Requests per Edge Node by Approach")
axs[1].set_xlabel("(b) Model")
axs[1].set_ylabel("Number of Assigned Requests")
axs[1].legend(title="Approach")
axs[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig("Results_perRequest.png", dpi=300, bbox_inches='tight')  # high quality
plt.show()
```
