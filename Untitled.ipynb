```python
import pandas as pd
import matplotlib.pyplot as plt


markers = ['o', 's', '^', 'v', '*']
y_fields = {
    "success_ratio": "Completed / Total Requests",
    "Speed:Deadline-Finish": "Response Time speed up",
    "Acc:AvgAcc-ExpAcc": "Accuracy improvement",
    "Total_energy_used": "Total Energy Used",
    "Total_utilisation": "Total Utilization"
}


arrival_log = pd.read_csv("log_diffArrv.csv")  # change path if needed
arrival_log["success_ratio"] = arrival_log["Completed_requests"] / (arrival_log["Completed_requests"] + arrival_log["Rejected_requests"])
approaches = arrival_log["approach"].unique()
fig, axs = plt.subplots(1, len(y_fields), figsize=(20, 4), sharey=False)
for i, (field, ylabel) in enumerate(y_fields.items()):
    ax = axs[i]
    for j, approach in enumerate(approaches):
        subset = arrival_log[arrival_log["approach"] == approach].sort_values("Arrival_rate")
        linewidth = 3 if approach == "Greedy" else 1.2
        ax.plot(
            subset["Arrival_rate"], subset[field],
            label=approach,
            marker=markers[j],
            linestyle='-', linewidth=linewidth
        )
    ax.set_xlabel("Arrival Rate")
    ax.set_ylabel(ylabel)
    ax.grid(True)
axs[0].legend(loc='best')
plt.tight_layout()
plt.show()


edge_log = pd.read_csv("log_diffEdge.csv") 
edge_log["success_ratio"] = edge_log["Completed_requests"] / (edge_log["Completed_requests"] + edge_log["Rejected_requests"])
approaches = edge_log["approach"].unique()
fig, axs = plt.subplots(1, len(y_fields), figsize=(20, 4), sharey=False)
for i, (field, ylabel) in enumerate(y_fields.items()):
    ax = axs[i]
    for j, approach in enumerate(approaches):
        subset = edge_log[edge_log["approach"] == approach].sort_values("edge_num")
        linewidth = 3 if approach == "Greedy" else 1.2
        ax.plot(
            subset["edge_num"], subset[field],
            label=approach,
            marker=markers[j],
            linestyle='-', 
            linewidth=linewidth
        )
    ax.set_xlabel("Number of edge servers")
    ax.set_ylabel(ylabel)
    ax.grid(True)
axs[0].legend(loc='best')
plt.tight_layout()
plt.show()


energy_log = pd.read_csv("log_diffEnergy.csv") 
energy_log["success_ratio"] = energy_log["Completed_requests"] / (energy_log["Completed_requests"] + energy_log["Rejected_requests"])
approaches = energy_log["approach"].unique()

unique_intervals = sorted(energy_log["Energy_interval"].unique())
interval_map = {interval: idx + 1 for idx, interval in enumerate(unique_intervals)}
energy_log["Energy_interval"] = energy_log["Energy_interval"].map(interval_map)

fig, axs = plt.subplots(1, len(y_fields), figsize=(20, 4), sharey=False)
for i, (field, ylabel) in enumerate(y_fields.items()):
    ax = axs[i]
    for j, approach in enumerate(approaches):
        subset = energy_log[energy_log["approach"] == approach].sort_values("Energy_interval")
        linewidth = 3 if approach == "Greedy" else 1.2
        ax.plot(
            subset["Energy_interval"], subset[field],
            label=approach,
            marker=markers[j],
            linestyle='-', 
            linewidth=linewidth
        )

    ax.set_xticks(list(interval_map.values()))
    ax.set_xticklabels(list(interval_map.keys()), rotation=45, ha='right')
    
    ax.set_xlabel("Energy limit of edge servers")
    ax.set_ylabel(ylabel)
    ax.grid(True)
axs[0].legend(loc='best')
plt.tight_layout()
plt.show()
```
